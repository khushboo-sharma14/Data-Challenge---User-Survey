# Apply NULL Hypothesis in order to check whether the result got from PIE Chart and line graph is correct or not.

### NULL Hypothesis can be thought of as the “control” of the experiment. The hypothesis assumed to be true before we collect data and usually has some sort of equal sign (≥, ≤, =).                                                                     Here we will use the p-value in order to evaluate our NULL Hypothesis. The p-value is the probability of observing a test statistic at least as large as the one observed, by random chance, given the null hypothesis is true.
### If p < α, we reject the null hypothesis. If p ≥ α, we fail to reject the null hypothesis.

```python
# Comparing with, No direct benefit/reputation for participating in social activities column
from scipy import stats
print(stats.ttest_ind(df["People will participate in social activities if their opinions are valued"], df["No direct benefit/reputation for participating in social activities"], equal_var=False))
```

***So, we fail to reject the NULL Hypothesis as our P-Value is more than 0.05***

### The same NULL Hypothesis We can apply to compare with other colums as well and in all the cases we would get the value more than 0.05

```python
# Example comparing with, People think that they don’t have something valuable and worthwhile to contribute to society. column
print(stats.ttest_ind(df["People will participate in social activities if their opinions are valued"], df["People think that they don’t have something valuable and worthwhile to contribute to society."], equal_var=False))
```

***Here also, we fail to reject the NULL Hypothesis as our P-Value is more than 0.05***
